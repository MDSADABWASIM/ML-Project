## Creating a custom Linear regression algorithm using pure python 

```python
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

### Create our data

```python
X = np.array([1,2,3,4,5,6], dtype=np.float64)
Y = np.array([5,4,6,5,6,7], dtype=np.float64)
```

### Getting the slope and intercept using y = mx + b (where m is slope and b is y intercept)

```python
def best_fit_slope_and_intercept(X,Y):
    m = (  ((mean(X) * mean(Y)) - mean(X*Y)) /
           ((mean(X) * mean(X)) - mean(X*X))  )

    b = mean(Y) - m * mean(X)
    return m, b
```

```python
m, b = best_fit_slope_and_intercept(X,Y)
print(m,b)
```

### Creating a regression line because now we have value of m(slope) and b(intercept)

```python
regression_line = [(m*x) + b for x in X]
```

### Plotting the regression line

```python
style.use('fivethirtyeight')
plt.scatter(X,Y)
plt.plot(X,regression_line)
plt.show()
```

### Making a prediction with this Linear model

```python
predict_X = 5
predict_Y = (m * predict_X) + b
print(predict_Y)
```

### Plotting our prediction over X

```python
plt.scatter(X,Y)
plt.scatter(predict_X,predict_Y, color='purple')
plt.plot(X,regression_line)
plt.show()
```

### Calculating the R-squared-error (Co-efficient of determination)

```python
def squared_error(ys_orig,ys_line):
    return sum((ys_line - ys_orig) * (ys_line - ys_orig))
```

```python
def coefficient_of_determination(ys_orig,ys_line):
    y_mean_line = [mean(ys_orig) for y in ys_orig]
    squared_error_regr = squared_error(ys_orig, ys_line)
    squared_error_y_mean = squared_error(ys_orig, y_mean_line)
    return 1 - (squared_error_regr/squared_error_y_mean)
```

```python
r_squared = coefficient_of_determination(Y,regression_line)
print(r_squared)
```

### Perfect R-square value will be 1.0
