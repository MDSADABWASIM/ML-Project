## Nifty 50 stock exchange data analysis project

The NIFTY 50 index is National Stock Exchange of India's benchmark stock market index for Indian equity market. It is a well diversified 50 stock index accounting for 22 sectors of the economy. It is used for a variety of purposes such as bench-marking fund portfolios, index based derivatives and index funds.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading the dataset

```python
df = pd.read_csv('../datasets/nifty50.csv',parse_dates=['date','time'])
df.head()
```

### Remove the index column

```python
df = df.drop('index',axis=1)
df.head(2)
```

```python
nifty_mean = df.groupby('date').mean()
nifty_mean.head(2)
```

### Visualization of nifty 50 data w.r.t date

```python
nifty_mean.plot(figsize=(12,8))
```

### The day nify 50 dipped huge

```python
maximum_up_nifty50 =  max(df['high'] - df['open'])
df.loc[(df['high'] - df['open']) == maximum_up_nifty50]
```

### The day nifty 50 raised huge

```python
maximum_drop_nifty =  max(df['open'] - df['low'])
maximum_drop_nifty
```

### Plotting hight values of nifty 50

```python
sns.set_style('darkgrid')
df['high'].plot('kde')
```

### Analysis of corelation 

```python
sns.heatmap(df.corr(),annot=True,cmap='magma')
```
